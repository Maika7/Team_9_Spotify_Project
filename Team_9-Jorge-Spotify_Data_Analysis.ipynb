```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Load the CSV file from the correct path
spotify_data_df = pd.read_csv("spotify_dataset.csv")

spotify_data_df.head()
```

```python
# Group by genre and calculate the mean popularity for each genre
genre_popularity = spotify_data_df.groupby('genre')['popularity'].mean().sort_values(ascending=False)

genre_popularity.head()
```

```python
# Get the top 10 genres by popularity
top_10_genres_by_popularity = genre_popularity.head(10)
top_10_genres_by_popularity.head()
```

```python
# Display the result
print("Top 10 Genres by Popularity:\n", top_10_genres_by_popularity)
```

```python
# 2. Calculate Top 10 Trending Genres in 2023

# Convert 'year' column to datetime (if it's not already in that format)
spotify_data_df['year'] = pd.to_datetime(spotify_data_df['year'], format='%Y')

spotify_data_df.head()
```

```python
# Filter the data for the year 2023
data_2023 = spotify_data_df[spotify_data_df['year'].dt.year == 2023]

data_2023.head()
```

```python
# Group by genre and calculate the average popularity in 2023
popularity_2023 = data_2023.groupby('genre')['popularity'].mean().reset_index()

popularity_2023.head()
```

```python

# Get the year 2022 data to calculate the change in popularity
data_2022 = spotify_data_df[spotify_data_df['year'].dt.year == 2022]

data_2022.head()
```

```python
popularity_2022 = data_2022.groupby('genre')['popularity'].mean().reset_index()

popularity_2022.head()
```

```python
# Merge 2023 and 2022 popularity data to calculate the difference
merged_data = pd.merge(popularity_2023, popularity_2022, on='genre', how='left', suffixes=('_2023', '_2022'))

merged_data.head()
```

```python
 #Calculate the change in popularity (popularity in 2023 minus popularity in 2022)
merged_data['popularity_change'] = merged_data['popularity_2023'] - merged_data['popularity_2022']

merged_data.head()
```

```python
# Sort by the change in popularity to get the trending genres
top_10_trending_genres_2023 = merged_data.sort_values(by='popularity_change', ascending=False).head(10)

top_10_trending_genres_2023.head()
```

```python
# Display the result
print("Top 10 Trending Genres in 2023:\n", top_10_trending_genres_2023[['genre', 'popularity_change']])
```

```python
# 3. Plot the Results

# Plot Top 10 Genres by Popularity
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_genres_by_popularity.values, y=top_10_genres_by_popularity.index)
plt.title('Top 10 Genres by Popularity (All Time)')
plt.xlabel('Average Popularity')
plt.ylabel('Genre')
plt.show()
```

```python
# Plot Top 10 Trending Genres in 2023
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_trending_genres_2023['popularity_change'], y=top_10_trending_genres_2023['genre'])
plt.title('Top 10 Trending Genres in 2023')
plt.xlabel('Popularity Change from 2022 to 2023')
plt.ylabel('Genre')
plt.show()
```
